```python
import numpy as np
import pandas as pd
# ML Pkgs
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
#from sklearn.metrics._plot import plot_confusion_matrix
  # Correct import statement
# DL Pkgs
from sklearn.neural_network import MLPClassifier
# DL tensorflow
# Load Dataset
dataset = pd.read_csv("bank-full.csv", sep=',')
# check size of dataset
dataset.shape
# check Datatype
dataset.dtypes
# Feature Engineering: Encode Dataset
# Get categorical
cat_columns = dataset.select_dtypes(include= ['object', 'category']).columns.tolist()
cat_columns = ['job', 'marital', 'education','default', 'housing','loan', 'contact', 'month', 'poutcome']
# Numerical Columns
num_columns = dataset.select_dtypes(exclude= ['object', 'category']).columns.tolist()
num_columns
# Label Encoding Target
dataset['Target'].unique()
# Label Encoding Target
dataset['Target'] = dataset['Target'].map({'no':0, 'yes':1})
# Feature Encoding : Categorical
ord_enc = OrdinalEncoder()
dataset[cat_columns] = ord_enc.fit_transform(dataset[cat_columns])
dataset[cat_columns]
# get Categories
ord_enc.categories_
# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
# Plot of Target Distribution
dataset['Target'].value_counts()
sns.countplot(x ='Target', data = dataset)
dataset['Target'].value_counts().div(dataset.shape[0]).mul(100)
# Distribution of Entire dataset
#plt.figure(figsize=(20, 10))
dataset.hist(figsize=(20, 10))
plt.show()
# Correlation Plot
corr_mat = dataset.corr()
corr_mat
plt.figure(figsize=(20, 10))
sns.heatmap(corr_mat, annot= True)
plt.show()
# Feature Selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
def select_best_features(xfeatures, ylabels, score_fxn = chi2):
  skb =SelectKBest(score_func= score_fxn, k = 10)
  best_feature_fit = skb.fit(xfeatures, ylabels)
  best_feat_dataset = pd.DataFrame({'feature_scores' :best_feature_fit.scores_, 'feature_name': xfeatures.columns})
  return best_feat_dataset
dataset.columns
Xfeatures = dataset[['age', 'job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign',
       'previous', 'poutcome']]
ylabels = dataset['Target']
Xfeatures.head(50)
best_features = select_best_features(Xfeatures, ylabels)
best_features
best_features.nlargest(12, 'feature_scores')
dataset.shape
best_features.nlargest(15, 'feature_scores')['feature_name'].tolist()
selected_features = best_features.nlargest(15, 'feature_scores')['feature_name'].tolist()
# Selected Features
Xselected = dataset[selected_features]
Xselected

# Split Dataset Before : Prevent Data Leakasge
X_train, X_test, y_train, y_test = train_test_split(Xselected, ylabels, test_size = 0.3, random_state = 42)
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = MinMaxScaler()
# Fit
sc.fit(X_train)
X_train_scaled = sc.transform(X_train)
# connect to df
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns.to_list())
X_train.head()
# scaled dataset
X_train_scaled.head()
# Apply on Test Data
X_test_scaled = sc.transform(X_test)
# convert to df
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns.to_list())
# Requirement
X_train_scaled.shape
print(y_train.unique())
input_dim = X_train_scaled.shape[1]
output_dim = 1
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import plot_model
# define model
model = Sequential()
model.add(Dense(20, activation='relu', kernel_initializer= 'he_normal', input_shape = (input_dim, )))
model.add(Dense(16, activation='relu', kernel_initializer= 'he_normal'))
model.add(Dense(output_dim, activation='softmax'))

def get_hidden_layers_neurons(sample_size, num_input, num_output, alpha = 2):
  num_hidden_layers = sample_size/(alpha * (num_input + num_output))
  return num_hidden_layers
# Compile Model
model.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
# Summary model
model.summary()
# Visualize the model
plot_model(model, to_file= 'bank_term_prediction_model.png', show_shapes=True)
# Fit the model
model.fit(X_train_scaled, y_train, epochs= 50, batch_size= 32, verbose= 2)
# Evaluate the model
loss, acc = model.evaluate(X_test_scaled, y_test, verbose= 0)
acc
loss
exl = X_test_scaled.iloc[0]
print(exl)
# Expected Value
y_test.iloc[0]
# predict
# Reshape dataset
sample1 = np.array(exl).reshape(1, -1)
# Probabilities
probabilities = model.predict(sample1)
probabilities
# Save Model
model.save('bank_term_predictor_model_2022.h5')
# Load Model
from tensorflow.keras.models import load_model
new_model = load_model('bank_term_predictor_model_2022.h5')
# Summary Model
new_model.summary()
history = model.fit(Xselected, ylabels, epochs= 50, batch_size= 32, verbose= 0, validation_split= 0.3)
history.history['loss']
# Plot
plt.title('Learning Curve')
plt.xlabel('Epochs')
plt.ylabel('Cross Entropy')
plt.plot(history.history['loss'], label = 'train')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()
# using Dropout on visible Input Layer
# define model
model2 = Sequential()
model2.add(Dropout(0.2, input_shape = (input_dim, )))
model2.add(Dense(20, activation='relu', kernel_initializer= 'he_normal'))
model2.add(Dense(16, activation='relu', kernel_initializer= 'he_normal'))
model2.add(Dense(output_dim, activation='softmax'))
# Compile Model
model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# Fit the Model
model2.fit(X_train_scaled, y_train, epochs = 50, batch_size = 32, verbose = 2)
# Evaluate the Model
loss2, acc2 = model2.evaluate(X_test_scaled, y_test, verbose=0)
print(loss2)
print(acc2)
# Deep Learning with Scikit-Learn
## MLPClassifier
from sklearn.neural_network import MLPClassifier
# Init
mlp_nn = MLPClassifier(max_iter= 50, random_state= 42)
mlp_nn.fit(X_train_scaled, y_train)
# accuracy
mlp_nn.score(X_test_scaled, y_test)
mlp_nn.predict(sample1)
```
